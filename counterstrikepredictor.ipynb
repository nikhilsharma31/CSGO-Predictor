```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder , RobustScaler
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
df= pd.read_csv('csgo_round_snapshots.csv')
```

```python
df.shape
```

```python
df.isnull().any()
```

```python
df.info()
```

```python
df.replace({'T':0 , 'CT':1} )
```

```python
df.drop(columns= 'map')
```

```python
df.replace({True:1  , False: 0} )
```

```python
df_copy=df.copy()
```

```python
dfc=df_copy
```

```python
dfc['bomb_planted']=dfc['bomb_planted'].map({True:1 , False:0})
```

```python
dfc['round_winner']= dfc['round_winner'].replace({'CT':1 , 'T' :0})
```

```python
dfc.head()
```

```python
dfc.drop('map', axis=1 , inplace= True)
```

```python
dfc
```

```python
y= dfc['round_winner']
X= dfc.drop('round_winner' , axis=1)
```

```python
X_train , X_test , y_train , y_test = train_test_split(X, y , test_size= 0.25)
```

```python
from sklearn.naive_bayes import GaussianNB , MultinomialNB
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score , precision_score , f1_score , recall_score
```

```python
lr=LogisticRegression()
```

```python
lr.fit(X_train , y_train)
lr_prediction = lr.predict(X_test)
lr_accuracy = accuracy_score(y_test , lr_prediction)
lr_f1score = f1_score(y_test , lr_prediction)
lr_recall_score = recall_score(y_test, lr_prediction)
lr_precision= precision_score(y_test , lr_prediction)
lr.score(X_test , y_test)
print('lr_precision=', lr_precision )
print('lr_recall+score =' , lr_recall_score)
print('lr_f1score = ' , lr_f1score )
print('lr_accuracy' , lr_accuracy)
```

```python
svm=SVC()
```

```python
svm.fit(X_train , y_train)
svm_prediction = svm.predict(X_test)
svm_accuracy = accuracy_score(y_test , svm_prediction)
svm_f1score = f1_score(y_test , svm_prediction)
svm_recall_score = recall_score(y_test , svm_prediction)
svm_precision = precision_score(y_test , svm_prediction)
svm.score(X_test, y_test)
print('accuracy = ' , svm_accuracy)
print('precision = ' , svm_precision)
print('recall=' , svm_recall_score)
print('f1score=' , svm_f1score)
```

```python
nb= GaussianNB()
nb.fit(X_train , y_train)
nb_prediction = nb.predict(X_test)
nb_accuracy = accuracy_score(y_test , nb_prediction)
nb_f1score = f1_score(y_test , nb_prediction)
nb_precision = precision_score(y_test , nb_prediction)
nb_recall = recall_score(y_test , nb_prediction)
nb.score(X_test , y_test)
print('accuracy=' , nb_accuracy)
print('precison=' , nb_precision)
print('recall=' , nb_recall)
print('f1score=' , nb_f1score)
```

```python
from sklearn.ensemble import RandomForestClassifier
import pandas as pd
X = X_scaled_df.drop(columns=['round_winner'])
y = X_scaled_df['round_winner']

# Create a Random Forest Classifier
rf = RandomForestClassifier()
rf.fit(X, y)

# Get feature importances
importances = rf.feature_importances_

# Create a DataFrame to store feature importances
feature_importances = pd.DataFrame({'Feature': X.columns, 'Importance': importances})

# Sort the features based on importance in descending order
top_features = feature_importances.nlargest(10, 'Importance')

# Print the top 5 features
print(top_features)
```
